### Handling Missing Values

```python
from sklearn.experimental import enable_iterative_imputer  # noqa
from sklearn.impute import IterativeImputer
from sklearn.ensemble import HistGradientBoostingRegressor
import pandas as pd
```

```python
df = pd.read_csv('weather_clean.csv', sep=';', parse_dates=['time'], index_col='time')
```

```python
df_numeric = df.select_dtypes(include='number')
```

#### Creating the Imputer that fills in missing values using RandomForestRegression

```python
# 1. Define a fast estimator
fast_tree = HistGradientBoostingRegressor(max_iter=100, random_state=42)
```

```python
# 2. Plug it into IterativeImputer
imputer = IterativeImputer(
    estimator=fast_tree,
    max_iter=5,
    random_state=42,
    verbose=2,
    initial_strategy='mean'
)
```

```python
# 3. Fit & transform in one go
df_imputed = pd.DataFrame(
    imputer.fit_transform(df_numeric),
    columns=df_numeric.columns,
    index=df_numeric.index
)

print("✅ Filled in all the missing values in record time!")
```

```python
# Merge data back to the categorical data (But we dont have it in our case)
# I added code just in case
df_final = df.copy()
df_final[df_numeric.columns] = df_imputed

# Saving results into the new file
df_final.to_csv('weather_final.csv', sep=';', index=True, encoding='utf-8-sig')
```
